```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('zomato_data_final.csv') 
df 
```

```python
df.isna().sum()
```

```python
import folium
from folium.plugins import MarkerCluster
from IPython.display import IFrame
```

```python
# Filter restaurants offering Italian cuisine
italian_df = df[df['cuisines'].str.contains("Italian", na=False)]

# Initialize Folium map centered on Bangalore
map4 = folium.Map(location=[12.9716, 77.5946], zoom_start=12)

# Add purple markers for Italian restaurants
for i, row in italian_df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['listed_incity']}<br>Cuisine: {row['cuisines']}",
        icon=folium.Icon(color='purple')
    ).add_to(map4)

# Save the map as HTML
map4.save("italian_spots_map.html")
```

```python
# Display the map in the notebook
IFrame('italian_spots_map.html', width=1000, height=800)
```
